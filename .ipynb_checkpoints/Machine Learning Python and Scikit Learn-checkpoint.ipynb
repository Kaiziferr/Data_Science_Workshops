# Preparación de los datos

Descripción del conjunto de datos: El conjunto de datos SaratogaHouses pertenece al paquete mosaicData de R. Contiene información de 1728 viviendas del año 2006.

Descripción de las caracteristicas:
- price: precio de la vivienda
- lotSize: metros cuadrados de la vivienda
- age: antiguedad de vivienda
- landValue: valor del terreno
- livingArea: metros cuadrados habitables
- pctCollege: porcentaje de vecindario con titulo universitario
- bedrooms: numero de dormitorios
- rooms: número de habitaciones
- heating: tipo de calefacción
- fuel: tipo de alimentación  de la calefacción (gas, electricidad o disel)
- sewer: tipo de desague
- waterfront: si la vivienda tiene vistas al lago
- newConstruction: si la vivienda es de nueva construcción
- centralAir: si la vivienda tiene aire acondicionado

```python
print(colored('Cargar datos', attrs=['bold']))
url = "https://raw.githubusercontent.com/JoaquinAmatRodrigo/Estadistica-machine-learning-python/master/data/SaratogaHouses.csv"
data = pd.read_csv(url, sep=',')
data.columns = ["precio", "metros_totales", "antiguedad", "precio_terreno",
                 "metros_habitables", "universitarios", "dormitorios", 
                 "chimenea", "cuartos_descanso", "habitaciones", "calefaccion",
                 "consumo_calefacion", "desague", "vistas_lago",
                 "nueva_construccion", "aire_acondicionado"]
data.head()
```

### Análisis inicial de nuestra variable objetivo

```python
fig, ax = plt.subplots(nrows = 3, ncols = 1, figsize = (6,6))
sns.distplot(
    data.precio,
    hist = False,
    rug = True,
    color = "b",
    kde_kws = {'shade': True, 'linewidth': 1},
    ax = ax[0]
)
ax[0].set_title('Distribución original', fontsize='medium')
ax[0].set_xlabel('Precio', fontsize='small')
ax[0].tick_params(labelsize =10)

sns.distplot(
    np.sqrt(data.precio),
    hist = False,
    rug = True,
    color = 'blue',
    kde_kws = {'shade':True, 'linewidth': 1},
    ax = ax[1]
)

ax[1].set_title('Distribución raiz cuadrada', fontsize='medium')
ax[1].set_xlabel('Precio', fontsize='small')
ax[1].tick_params(labelsize =10)

sns.distplot(
    np.log(data.precio),
    hist=False,
    rug=True,
    color = 'blue',
    kde_kws = {'shade':True, 'linewidth': 1},
    ax = ax[2]
)

ax[2].set_title("Transformación logarítmica", fontsize = 'medium')
ax[2].set_xlabel('log(precio)', fontsize='small') 
ax[2].tick_params(labelsize = 10)

fig.tight_layout()
```



Algunos modelos de machine learning y aprendizaje estadístico requieren que la variable respuesta se distribuya de una forma determinada. Por ejemplo, para los modelos de regresión lineal (LM), la distribución tiene que ser de tipo normal. Para los modelos lineales generalizados (GLM), la distribución tiene que ser de la familia exponencial.

Existen varias librerías en python que permiten identificar a qué distribución se ajustan mejor los datos, una de ellas es fitter. Esta librería permite ajustar cualquiera de las 80 distribuciones implementadas en scipy.


```python
distribuciones = ['cauchy', 'chi2', 'expon',  'exponpow', 'gamma',
                  'norm', 'powerlaw', 'beta', 'logistic']
fitter = Fitter(data.precio, distributions = distribuciones)
fitter.fit()
fitter.summary(Nbest=10)
```

```python
datos_numerics = (data.dtypes == 'int64') | (data.dtypes == 'float64')
datos_categorics = (data.dtypes == 'object')
datos_numerics = [c for c in datos_numerics.index if datos_numerics[c]]
datos_categorics = [c for c in datos_categorics.index if datos_categorics[c]]
datos_numerics = datos_numerics[1:]
```

```python
fig, ax = plt.subplots(nrows = 3, ncols = 3, figsize = (19,15))
ax = ax.flat
for i, colunm in enumerate(datos_numerics):
    sns.distplot(
        data[colunm], 
        hist = True, 
        rug = False, 
        color   = (list(plt.rcParams['axes.prop_cycle'])*2)[i]["color"],
        kde_kws = {'shade': False, 'linewidth': 1},
        ax      = ax[i])
    
    ax[i].set_title(colunm, fontsize = 17, fontweight = "bold")
    ax[i].tick_params(labelsize = 16)
    ax[i].set_xlabel("")

    
for i in [8]:
    fig.delaxes(ax[i])

fig.tight_layout()
plt.subplots_adjust(top=0.9)
fig.suptitle('Correlación con precio', fontsize = 20, fontweight = "bold");
    
fig.tight_layout()
plt.subplots_adjust(top = 0.9)
fig.suptitle('Distribución variables numéricas', fontsize = 20, fontweight = "bold");
```

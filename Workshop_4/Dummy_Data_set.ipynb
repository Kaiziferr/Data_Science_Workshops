# Dummy Data set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000000
data = pd.DataFrame({
    'A': np.random.randn(n),
    'B': 1.5 + 2.5 * np.random.randn(n),
    'C': np.random.uniform(5,32,n)
})
```

```python
data.head()
```

```python
plt.hist(data.A, bins = 100)
plt.axvline(data.A.mean(), c='r', linestyle = '--', label = 'Promedio')
plt.axvline(data.A.std(), c='b', linestyle = '--', label = 'Desviiacion +')
plt.axvline(-data.A.std(), c='b', linestyle = '--', label = 'Desviiacion -')
plt.legend()
```

```python
plt.hist(data.B, bins = 100)
plt.axvline(data.B.mean(), c='r', linestyle = '--', label = 'Promedio')
plt.axvline(1.5 + 2.5*data.B.std(), c='b', linestyle = '--', label = 'Desviiacion +')
plt.axvline(1.5 - 2.5*data.B.std(), c='b', linestyle = '--', label = 'Desviiacion -')
plt.legend()
```

#### Agregación de datos por categoría

```python
gender = ['Male','Female']
income = ['Poor','Middle','Rich']
```

```python
n = 500
gender_data = []
income_data = []
for i in range(0,500):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
gender_data[0:5]
```

```python
income_data[0:5]
```

```python
#Los valores, no necesariamente deben ser acorde a la realidad, esto es con fin para mostrar el ejercicio
height = 175+25*np.random.randn(n)
weight = 65 + 25*np.random.randn(n)
age = 30+12*np.random.randn(n)
income = 18000+3500*np.random.rand(n)
```

```python
data = pd.DataFrame({
    "Gender": gender_data,
    "Economic_Status":income_data,
    "Icome": income,
    "Height": height,
    "Weight": weight,
    "Age":age
})
```

```python
data.head()
```

#### Agrupacion

```python
grouped_gender = data.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
for names, groups in grouped_gender:
    print(names, groups)
```

```python
grouped_gender.get_group('Male').head()
```

```python
double_group = data.groupby(["Gender","Economic_Status"])
```

```python
len(double_group)
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

```python
double_group.aggregate(
    {
        "Icome": np.sum,
        "Age":np.mean,
        "Height":np.std
    }
)
```

```python
double_group.aggregate(
    {
        "Age":np.mean,
        "Height": lambda h:(np.mean(h)/np.std(h))
    }
)
```

```python
double_group.aggregate([np.sum,np.mean, np.std])
```

```python
double_group.aggregate([lambda x: np.mean(x)/np.std(x)])
```

#### Filtrado

```python
double_group["Age"].filter(lambda x:x.sum()>2400)
```

#### Transformacion de variables

```python
zscore = lambda x: (x-x.mean())/x.std()
```

```python
z_score = double_group.transform(zscore)
```

```python
plt.hist(z_score['Age'],bins=30)
```

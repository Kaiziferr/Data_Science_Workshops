# Merge

```python
#Eliminar atletas que participaron por varios paises
#No se con que finalidad, por que se perderia valor, pero como son pocos datos, se podria sin afectar el conjunto de datos
#Solo conservar el conjunto de Athlete unicos
data_country_dp = data_country.drop_duplicates(subset = "Athlete")
```

```python
data_main_country = pd.merge(left=data_main, right = data_country_dp, left_on = "Athlete", right_on="Athlete")
```

```python
data_main_country.head()
```

```python
len(data_main_country)
```

```python
data_main_country[data_main_country["Athlete"] == "Aleksandar Ciric"]
```

- En el tutorial eliminan los registros en el cual el atleta se repite pero con otro deporte, si bien es necesario para hacer un join, se arriesga a perder información relevante.

```python
data_sport_dp = data_sport.drop_duplicates(subset = "Athlete")
len(data_sport_dp)
```

```python
data_final = pd.merge(left=data_main_country, right=data_sport_dp, left_on = "Athlete", right_on="Athlete")
data_final.head()
```

```python
### Modificar los conjuntos de datos para ver el efecto de los joins
out_athletes = np.random.choice(data_main['Athlete'], size = 6, replace=False)
```

```python
data_country_dit = data_country_dp[~(data_country_dp['Athlete'].isin(out_athletes)) & (data_country_dp['Athlete'] != 'Michael Phelps')]
data_sport_dit = data_sport_dp[~(data_sport_dp['Athlete'].isin(out_athletes)) & (data_sport_dp['Athlete'] != 'Michael Phelps')]

data_main_dit = data_main[~(data_main['Athlete'].isin(out_athletes)) & (data_main['Athlete'] != 'Michael Phelps')]
```

```python
data_sport_dit.head()
```

```python
data_main_dit.head()
```

```python
display(len(data_country_dit), len(data_sport_dit), len(data_main_dit))
```

# Inner Join (Interseccion)

```python
merged_inner = pd.merge(left=data_main, right=data_country_dit, how="inner", left_on = "Athlete", right_on = "Athlete")
merged_inner.head()
```

```python
len(merged_inner)
```

# Left Join (La info de el dataset A y la interseccion de B, excluyendo el resto de B)

```python
merged_left = pd.merge(left = data_main, right = data_country_dit, how = "left",left_on="Athlete", right_on="Athlete")
len(merged_left)
```

```python
merged_left.head()
```

# Rigth join

```python
merged_rigth = pd.merge(left = data_main_dit, right = data_country_dp, how = "right",left_on="Athlete", right_on="Athlete")
len(merged_rigth)
```

```python
merged_rigth.tail(10)
```

# Outer Join

```python
data_country_jb = data_country_dit.append({
    "Athlete": "Steve Bernal",
    "Country": "Colombia"
}, ignore_index = "True")
```

```python
merged_outer = pd.merge(left = data_main, right = data_country_jb, how="outer", left_on="Athlete", right_on="Athlete")
len(merged_outer)
```
